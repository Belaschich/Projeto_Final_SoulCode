# TRATATIVA BASE AMÉRICA DO NORTE

```python
#DROP DE COLUNAS IRRELEVANTES 

df1.drop('PerCapita3yr', axis=1, inplace=True)
df1.drop('PctChange', axis=1, inplace=True)
```

```python
#DROP DE LINHAS NULAS

df1.dropna(axis=0, inplace = True)
```

```python
#TRADUÇÃO DE COLUNAS

df1.rename(columns={'Year': 'Ano','Month': 'Mes', 'Beverage': 'Categoria', 'Gallons':'Galoes', 'Ethanol': 'Etanol' , 'Population': 'Populacao', 'FIPS': 'Estado', 'PerCapita': 'EtanolPerCapita'}, inplace=True)
```

```python
#TRADUÇÃO DE ESTADOS *INFORMADO NA FONTE DE CONSULTA* (FISP para NomeEstado)

df1['Estado'].replace(2, 'Alaska', inplace=True)
df1['Estado'].replace(8, 'Colorado', inplace=True)
df1['Estado'].replace(9, 'Connecticut', inplace=True)
df1['Estado'].replace(10, 'Delaware', inplace=True)
df1['Estado'].replace(12, 'Florida', inplace=True)
df1['Estado'].replace(17, 'Illinois', inplace=True)
df1['Estado'].replace(21, 'Kentucky', inplace=True)
df1['Estado'].replace(25, 'Massachusetts', inplace=True)
df1['Estado'].replace(27, 'Minnesota', inplace=True)
df1['Estado'].replace(29, 'Missouri', inplace=True)
df1['Estado'].replace(38, 'Dakota do Norte', inplace=True)
df1['Estado'].replace(47, 'Tennessee', inplace=True)
df1['Estado'].replace(48, 'Texas', inplace=True)
```

```python
#VALIDAÇÃO COLUNA "ESTADO"

df1['Estado'].unique()
```

```python
#TRADUÇÃO DE CATEGORIA *INFORMADO NA FONTE DE CONSULTA*

df1['Categoria'].replace(1, 'Destilado', inplace=True)
df1['Categoria'].replace(2, 'Vinho', inplace=True)
df1['Categoria'].replace(3, 'Cerveja', inplace=True)
df1['Categoria'].replace(4, 'Todos', inplace=True)
```

```python
#VALIDAÇÃO COLUNA CATEGORIA

df1['Categoria'].unique()
```
